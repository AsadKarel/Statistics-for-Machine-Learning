# Kurtosis values and their features

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

----------------

```python
data = pd.read_csv('D:/PGP_DSE/NPV/titanic.csv')
data.head(2)
```

```python
a = data.Fare.dropna()
b = np.sqrt(a) ; c = np.sqrt(b) ; d = np.sqrt(c) ; e = np.sqrt(d)
```

--------------------

## Distribution for population data:

```python
sns.distplot(a, hist=False)
plt.show()
```

-----------------

## Normalization:

```python
sns.distplot(b, hist=False)
plt.show()
```

```python
sns.distplot(c, hist=False)
plt.show()
```

```python
sns.distplot(e, hist=False)
plt.show()
```

-------------------

## Cheking of Kurtosis with all distributions:

```python
sns.distplot(b, hist=False, color='Black', label = 'b = 6.29 , skew = 2.08')
sns.distplot(c, hist=False, color='green', label = 'c = 2.88 , skew = 0.51')
sns.distplot(d, hist=False, color='red', label = 'd = 13.97, skew = -2.35')
#sns.distplot(e, hist=False, color='orange', label = 'e = 36.31')
plt.show()
```

--------------------------

```python
ab = a.kurt() , b.kurt(), c.kurt() , d.kurt() , e.kurt() , np.sqrt(e).kurt() , np.sqrt(np.sqrt(e)).kurt() , np.sqrt(np.sqrt(np.sqrt(e))).kurt() , np.sqrt(np.sqrt(np.sqrt(np.sqrt(e)))).kurt() , np.sqrt(np.sqrt(np.sqrt(np.sqrt(np.sqrt(e))))).kurt(), np.sqrt(np.sqrt(np.sqrt(np.sqrt(np.sqrt(np.sqrt(e)))))).kurt() , np.sqrt(np.sqrt(np.sqrt(np.sqrt(np.sqrt(np.sqrt(np.sqrt(e))))))).kurt()
ab
```

## Cheking exsiliency of Kurtosis values along Normalizations:

```python
sns.barplot(x=np.arange(1,13), y = ab)
plt.xlabel('No of observations.')
plt.ylabel('Kurtosis')
plt.show()
```

```python
sns.lineplot(x=np.arange(1,13), y = ab, color='black')
plt.xlabel('No of observations.')
plt.ylabel('Kurtosis')
plt.show()
```
